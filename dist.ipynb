```python
import matplotlib.pyplot as plt, numpy as np

x = np.arange(0.0, 2*np.pi, 0.01)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt, numpy as np

groups, scores = np.arange(7), (5, 8, 15, 17, 12, 8, 5)
plt.bar(groups, scores)
plt.title('Some Data')
plt.xlabel('x again')
plt.ylabel('y again')
plt.show()
```
