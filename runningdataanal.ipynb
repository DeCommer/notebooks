```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
boobs = pd.read_csv('/Users/josephdecommer/Documents/coding2022/Python/Data Sets/runningdata.csv')
```

```python
boobs.shape
```

```python
boobs.head()
```

```python
boobs['log_distance'] = np.log(boobs['distance'])
boobs['log_est_calories'] = np.log(boobs['est_calories'])
```

```python
plt.scatter(boobs['distance'], boobs['est_calories'])
plt.xlabel('Distance')
plt.ylabel('Est Calories')
```

```python
plt.scatter(boobs['year'], boobs['est_calories'])
plt.xlabel('Year')
plt.ylabel('Est Calories')
```

```python
plt.scatter(boobs['log_distance'], boobs['est_calories']) # x scaled
plt.xlabel('Distance as Log')
plt.ylabel('Estimated Calories')
```

```python
plt.scatter(boobs['distance'], boobs['log_est_calories']) # y scaled
plt.xlabel('Distance')
plt.ylabel('Estimated Calories as Log')
```

```python
plt.scatter(boobs['log_distance'], boobs['log_est_calories']) #both x and y scaled
plt.xlabel('Distance as Log')
plt.ylabel('Estimated Calories as Log')
```
